```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
import warnings; warnings.simplefilter('ignore')
```

```python
iris_data = load_iris()
for i in iris_data.keys():
    print(iris_data['DESCR'])
```

```python
dataset = pd.DataFrame(
    data= np.c_[iris_data['data'], iris_data['target']],
    columns=iris_data['feature_names'] + ['target']
)
```

```python
print(f'Размерность датасета: {dataset.shape}')
print(f'Колонки: {dataset.columns}')
dataset.dtypes
```

```python
import seaborn as sns
%matplotlib inline
sns.set(style="ticks")
```

sepal - чашелистик
petal - лепесток

Из графиков видно, что у ирисов Фишера есть почти линейная зависимость между следующими парами параметров:
* petal_length : petal_width
* sepal_length : petal_length

Также можно заметить сильновыраженную классификацию по габаритам чашелистика

```python
sns.pairplot(dataset)
```

Рассмотрим также диаграммы распределения по существующим классам

```python
sns.pairplot(dataset, hue='target')
```
